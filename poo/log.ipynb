```python
class App():
    def __init__(self, user, capacidad, name):
        self.usuario=user
        self.capa=capacidad
        self.nombre=name
    def login(self):
        if self.nombre=="Dueño" and self.usuario>=1:
            print(f"bienvenido {self.nombre}")
            print(f"Tu capacidad es de {self.capa}")
    def increm(self, num):
        self.capa+=num
        return self.capa

us1=App(35,256,"Dueño")
us1.login()
us1.increm(56)

us2=App(40,245,"Alan")
us2.login()
```
