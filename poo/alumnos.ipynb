```python
class Alumno():
    def __init__(self, nom, matr):
        self.nombre=nom
        self.matri=matr
        self.calif=[]
    def ag_calif(self, califica):
        self.calif.append(califica)
    def cprom(self):
        return sum(self.calif)/len(self.calif)
    def final(self):
        promedio=self.cprom()
        return "Aprobado" if promedio>=7 else "Reprobao"
    
class Grupo():
    def __init__(self, nombre):
        self.nomg=nombre
        self.alumnos=[]
    def agregar(self, alumno):
        self.alumno=alumno
    def mostrar(self):
        print(f"Promedios del grupo {self.nomg}: ")
        for alumno in self.alumnos:
            promedio=alumno.cprom()
            estado=alumno.final()
            print(f"{alumno.nombre} ({alumno.matri}): Promedio = {promedio: .2f}, Estado = {estado}")
    def best(self):
        if not self.alumnos:
            print("No hay alumnos en el grupo")
            return None
        mejor= max(self.alumnos, key=lambda alumno: alumno.cprom())
        print(f"El mejor alumno es {mejor.nombre} ({mejor.matri}) con un promedio de {mejor.cprom(): .2f}")

        
  
```

```python
alu1=Alumno("Andree", 4232445)
alu2=Alumno("Osvaldo", 534534)
alu3=Alumno("Majo", 457474)
```

```python
alu1.ag_calif(9.5)
alu1.ag_calif(8)
alu2.ag_calif(10)

alu2.ag_calif(10)
alu2.ag_calif(10)

alu3.ag_calif(6)
alu3.ag_calif(8)
alu3.ag_calif(9)
alu3.ag_calif(7)
```

```python
print(f"El alumno {alu1.nombre} tiene promedio de: {alu1.cprom()} y esta: {alu1.final()}")
print(f"El alumno {alu2.nombre} tiene promedio de: {alu2.cprom()} y esta: {alu2.final()}")
print(f"El alumno {alu3.nombre} tiene promedio de: {alu3.cprom()} y esta: {alu3.final()}")
```
